```python
import os
import csv

def find_log(path):
    isfind = False
    log = None
    for a in os.listdir(path):
        if a.endswith('.log'):
            log = a
            isfind = True
    return isfind, log

def read_dir(filepath):
    path = []
    name = []
    for file in os.listdir(filepath): 
        if file == "test":
            continue
        subpath = os.path.join(filepath, file)
        
        isfind, log = find_log(subpath)

        if isfind:
            name.append(file)
            path.append(os.path.join(subpath, log))

    return name, path

def line_split(line):
    line_temp = line.rstrip('\n')
    line_temp = line_temp.split(':')[1]
    line_temp = line_temp.split(', ')

    return line_temp

def read_data(name, path, metrics, objects, data_dict):
    line_count = 0
    #print(path)
    #data_dict = {'tag': name}
    with open(path) as file:
        lines = file.readlines()
        data_car_3d = {'easy': 0.0, 'mod': 0.0, 'hard': 0.0}
        data_car_bbox = {'easy': 0.0, 'mod': 0.0, 'hard': 0.0}
        data_car_bev = {'easy': 0.0, 'mod': 0.0, 'hard': 0.0}
        data_ped_3d = {'easy': 0.0, 'mod': 0.0, 'hard': 0.0}
        data_ped_bbox = {'easy': 0.0, 'mod': 0.0, 'hard': 0.0}
        data_ped_bev = {'easy': 0.0, 'mod': 0.0, 'hard': 0.0}
        data_cyclist_3d = {'easy': 0.0, 'mod': 0.0, 'hard': 0.0}
        data_cyclist_bbox = {'easy': 0.0, 'mod': 0.0, 'hard': 0.0}
        data_cyclist_bev = {'easy': 0.0, 'mod': 0.0, 'hard': 0.0}
        for line in lines:
            #print(line)
            car = line.find('AP_R40@0.70, 0.70, 0.70:')
            #car
            if car > 0:
                #print('find ', name)

                #bbox
                if 'bbox' in metrics:
                    line_temp = line_split(lines[line_count+1])
                    if float(line_temp[1]) > data_car_bbox['mod']:
                        data_car_bbox = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                        if 'ped' in objects:
                            line_temp = line_split(lines[line_count+21])
                            data_ped_bbox = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                        if 'cyclist' in objects:
                            line_temp = line_split(lines[line_count+41])
                            data_cyclist_bbox = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                #bev
                if 'bev' in metrics:
                    line_temp = line_split(lines[line_count+2])
                    if float(line_temp[1]) > data_car_bev['mod']:
                        data_car_bev = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                        if 'ped' in objects:
                            line_temp = line_split(lines[line_count+22])
                            data_ped_bev = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                        if 'cyclist' in objects:
                            line_temp = line_split(lines[line_count+42])
                            data_cyclist_bev = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                #3d
                if '3d' in metrics:
                    line_temp = line_split(lines[line_count+3])
                    if float(line_temp[1]) > data_car_3d['mod']:
                        data_car_3d = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                        if 'ped' in objects:
                            line_temp = line_split(lines[line_count+23])
                            data_ped_3d = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                        if 'cyclist' in objects:
                            line_temp = line_split(lines[line_count+43])
                            data_cyclist_3d = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}

                # #car_3d
                # line_temp = line_split(lines[line_count+3])
                # if float(line_temp[1]) > data_car_3d['mod']:
                #     data_car_3d = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                
                #     line_temp = line_split(lines[line_count+2])
                # #if float(line_temp[1]) > data_car_bev['mod']:
                #     data_car_bev = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}

                #     line_temp = line_split(lines[line_count+1])
                # #if float(line_temp[1]) > data_car_bbox['mod']:
                #     data_car_bbox = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
            
                # #ped
                # line_temp = line_split(lines[line_count+21])
                # if float(line_temp[1]) > data_ped_bbox['mod']:
                #     data_ped_bbox = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}

                #     line_temp = line_split(lines[line_count+22])
                # #if float(line_temp[1]) > data_ped_bev['mod']:
                #     data_ped_bev = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}

                #     line_temp = line_split(lines[line_count+23])
                # #if float(line_temp[1]) > data_ped_3d['mod']:
                #     data_ped_3d = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                
                # #cyclist
                # line_temp = line_split(lines[line_count+41])
                # if float(line_temp[1]) > data_cyclist_bbox['mod']:
                #     data_cyclist_bbox = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}

                #     line_temp = line_split(lines[line_count+42])
                # #if float(line_temp[1]) > data_cyclist_bev['mod']:
                #     data_cyclist_bev = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}

                #     line_temp = line_split(lines[line_count+43])
                # #if float(line_temp[1]) > data_cyclist_3d['mod']:
                #     data_cyclist_3d = {'easy': float(line_temp[0]), 'mod': float(line_temp[1]), 'hard': float(line_temp[2])}
                    
                line_count += 1

            else:
                line_count += 1


    dict_temp = {'car_bbox': data_car_bbox, 'car_bev': data_car_bev, 'car_3d': data_car_3d,
                     'ped_bbox': data_ped_bbox, 'ped_bev': data_ped_bev, 'ped_3d': data_ped_3d, 
                     'cyclist_bbox': data_cyclist_bbox, 'cyclist_bev': data_cyclist_bev, 'cyclist_3d': data_cyclist_3d, }
    data_dict[name] = dict_temp
    return data_dict

def write_csv(method, metrics, data_dict):
    csv_file = open('result/' + method + '_result.csv', 'w', newline='', encoding='gbk')
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['', '', '', 'car', '', '', '', 'ped', '', '', '', 'cyclist', '', ''])
    csv_writer.writerow(['', 'tag', '', 'easy', 'mod', 'hard', '', 'easy', 'mod', 'hard', '', 'easy', 'mod', 'hard'])
    for metric in metrics:
        csv_writer.writerow([metric])
        for key in data_dict:
            csv_writer.writerow(['', key, '', data_dict[key]['car_' + metric]['easy'], data_dict[key]['car_' + metric]['mod'], data_dict[key]['car_' + metric]['hard'], 
                '', data_dict[key]['ped_' + metric]['easy'], data_dict[key]['ped_' + metric]['mod'], data_dict[key]['ped_' + metric]['hard'], 
                '', data_dict[key]['cyclist_' + metric]['easy'], data_dict[key]['cyclist_' + metric]['mod'], data_dict[key]['cyclist_' + metric]['hard'], ])
        csv_writer.writerow([])

    print('write down!')
    csv_file.close()





    
```
